```python
import spacy

nlp = spacy.load('en_core_web_sm')
```

```python
def POS_tagging(rows):
    try:
        document_words = rows["Word"]
        result = u' '.join(list(document_words))
        tagged = nlp(result)
        count = 0
        
        for token in tagged:
            rows.loc[rows['Word'] == token.text, 'pos_tag'] = token.tag_
    except:
        pass
    
    return rows
```

```python
class SentenceGetter(object):
    
    def __init__(self, data):
        self.sent_index = 1
        self.data = data
        self.empty = False
        
        aggregate_func = lambda x: [(w, p, t) for w, p, t in zip(x['Word'].values.tolist(), x['pos_tag'].values.tolist(), x['tag'].values.tolist())]
            
        self.grouped = self.data.groupby('Sent_ID').apply(aggregate_func)
        self.sentences = [s for s in self.grouped]
            
    def get_next(self):
        sent = self.grouped[self.sent_index]
        self.sent_index += 1
        return sent
```

```python
def word_2_features(sentence, word_index):
    
    word = sentence[word_index][0]
    pos_tag = sentence[word_index][1]
    
    # static features
    
    features = {
        'bias' : 1,
        'word[-3:]' : word[-3:],
        'word[-2:]' : word[-2:],
        
        'lower' : word.islower(),
        'upper' : word.isupper(),
        'digit' : word.isdigit(),
        'hyphen' : word.count('-'),
        'title' : word.istitle(),
        
        'postag' : pos_tag,
        'postag2' : pos_tag[:2],

        '#capitals' : sum(1 for c in word if c.isupper()),
    }
    
    # transition features
    
    # prev
    if word_index > 0:
        prev = sentence[word_index-1][0]
        prev_pos_tag = sentence[word_index-1][0]
         
        features.update({
            'prev_lower' : prev.islower(),
            'prev_upper' : prev.isupper(),
            'prev_title' : prev.istitle(),
            
            'prev_postag' : prev_pos_tag,
            'prev_postag2' : prev_pos_tag[:2]
        })    
    
    else:
        features['BOS'] = True
        
    if word_index < (len(sentence) - 1):
        nex = sentence[word_index+1][0]
        nex_pos_tag = sentence[word_index+1][1]
        
        features.update({
            'nex_lower' : nex.islower(),
            'nex_upper' : nex.isupper(),
            'nex_title' : nex.istitle(),
            
            'nex_postag' : nex_pos_tag,
            'nex_postag2' : nex_pos_tag[:2]
        })
        
    else:
        features['EOS'] = True
        
    return features
```

```python
def get_labels(sentence):
    return [t for w,p,t in sentence]
```

```python
def get_features(sentence):
    return [word_2_features(sentence, i) for i in range(len(sentence))]
```
